Conditional Statement

```python
num = int(input())
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")    
```

Loop and list

```python
numbers = input().split()
even_sum = 0
even_count = 0

for num in numbers:
    num = int(num)
    if num % 2 == 0:
        even_sum += num
        even_count += 1
print("Average of even numbers:", even_sum / even_count)
```

```python
numbers = list(map(int, input().split(" ")))
even_sum = 0
even_count = 0

for num in numbers:
    num = int(num)
    if num % 2 == 0:
        even_sum += num
        even_count += 1
print("Even Sum:", even_sum)
print("Even Count:", even_count)    
```

Dictionary

```python
numbers = input().split()

freq = {}
for num in numbers:
    num = int

    if num in freq:
        freq[num] += 1

    else:
        freq[num] = 1
print(freq)
```

Generator<br>Yield

```python
def data_loader(batch_size):
    for i in range(0, len(df), batch_size):
        yield df.iloc[i:i+batch_size]#e
data = np.arange(10)
loader = data_loader(5,data)
print(next(loader))
print(next(loader))
```

Filter

```python
import numpy as np
import pandas as pd       

data =np.array([10, 20, 30, 40, 50])
filtered_np = list (filter(lambda x: x > 25, data))
print(filtered_np)
```
