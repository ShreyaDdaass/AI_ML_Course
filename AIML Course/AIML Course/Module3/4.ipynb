File Pointer/ Cursor

```python
file = open("example.txt", "r")
print (file.tell())# it will return the current position of the file pointer or the cursor
print (file.read(74))# read 74 characters
print (file.tell())
```

```python
# Read
file = open("example.txt", "r")
print (file.tell())
print (file.read(5))
print (file.tell())

print (file.read())
```

```python
# Seek
file = open("example.txt", "r")
print (file.tell())
file.seek(21)
print (file.tell())
```
