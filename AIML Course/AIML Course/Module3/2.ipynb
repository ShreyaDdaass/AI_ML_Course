File Read and Write

```python
file = open("example.txt", "r")
content = file.read()
print (content)
print (type(content))
```

```python
file = open("example.txt", "r")
content = file.readlines()
print (content)
print (type(content))
```

```python
file = open("example.txt", "r")
content = file.readlines()
print (content)
print (type(content))
print(file.closed)
file.close()
# print(file.closed)
```

```python
with open("example.txt", "r") as file:
    content = file.read()
    print (content)
print (file.closed)
```

```python
with open("example.txt", "r") as file:
    for line in file:
        print (line)
```

```python
with open("example.txt", "r") as file:
    for line in file:
        print (line,end = "")
```

```python
with open("example.txt", "r") as file:
    for line in file:
        l = line.strip() # to remove the newline character
        print (l)
```
