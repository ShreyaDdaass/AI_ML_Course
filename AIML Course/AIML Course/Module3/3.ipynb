File Write and Append

```python
# Creating a new file and writing there

with open ("new.txt", "w") as file:
    file.write("Hello, this is an example file.\n")
    file.write("It contains multiple lines of text.\n")
    file.write("This is the third line.\n")
```

```python
# writing on a existing file
# It got over writen
with open ("new.txt", "w") as file:
    file.write("Hello,so you cannot see me.\n")
    file.write("You cannot see me.\n")
    file.write("Because you are looking at the wrong place.\n")
```

```python
# no over writing, but appending
with open ("new.txt", "a") as file:
    file.write("Hello,so you cannot see me ,you do not know .\n")
    file.write("You buy me a chocolate.\n")
    file.write("If you can not then get lost.\n")
```
