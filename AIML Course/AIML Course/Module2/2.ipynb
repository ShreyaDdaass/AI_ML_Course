Structure of a function - Parameters

```python
def welcome_message():
    print ("Hello User")#no parameter or return type function
    # we need to call it to run
welcome_message()# calling the function
```

```python
def welcome_message(User):
    print (f"Hello {User}")
welcome_message("Alice")
```

```python
def welcome_message(User = "None"):
    print (f"Hello {User}")
welcome_message("Alice")
```
