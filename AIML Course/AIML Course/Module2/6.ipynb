Return type of a function

```python
def welcome_message():
    print ("Hello User")
print (type(welcome_message()))
```

```python
def adder(a,b):
    sum = a+b
    return sum
print (type(adder(2,4)))
```

```python
def adder(a,b):
    sum = a+b
    return sum
print (type(adder(2.32,4.897)))
```

```python
# default parameter value

def welcome_message(User = "None"):
    print (f"Hello {User}")
print (type(welcome_message()))
```

```python
def welcome_message(User = "None"):
    return f"Hello {User}"
print (type(welcome_message()))
```

```python
def adder(a,b):
    return a,b
print (type(adder(2,4)))

# unzip
x,y = adder(2,4)
print (x)
print (y)
```
