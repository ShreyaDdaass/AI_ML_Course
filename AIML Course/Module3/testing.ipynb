```python
file = open("data.txt", "r")
with open("data.txt", "r") as file:
    text = file.read()

# Convert to lowercase
text = text.lower()

# Remove punctuation
for p in string.punctuation:
    text = text.replace(p, "")

# Split into words
words = text.split()

# Count frequency
freq = {}

for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

# Find most frequent word
max_word = max(freq, key=freq.get)

print(max_word)
```
