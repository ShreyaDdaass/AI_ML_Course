Iterator and Generator

Iterator

```python
s = {10, 20, 30, 40, 50}
print (s[1])
```

```python
s = {10, 20, 30, 40, 50}
s__iter = iter(s)
print (s__iter)
```

```python
s = {10, 20, 30, 40, 50}
s__iter = iter(s)
print (next(s__iter))
print (next(s__iter))
print (next(s__iter))
print (next(s__iter))
print (next(s__iter))  

for i in s:
    print (i)   
```

Generator

```python
list =[x for x in range(10)]#dataset
def data_loader(chunk_size,list):
    for i in range(0,len(list),chunk_size):
        yield list[i:i+chunk_size]
x = data_loader(3,list)
print (next(x))
print (next(x))

# only loading the part it needs not the whole 10 numbers at once. This is memory efficient and faster.
```
