Keyword Arguments

```python
def adder (a,b,c):
    print (f"{a},{b},{c}")

adder (a=2,b=4,c=3) 
```

```python
def adder (**kwargs):
    print (kwargs)
adder (a=2,b=4,c=3,  n_estimator=50 )
```

```python
def adder (**kwargs):
    for key, value in kwargs.items():
        print (f"{key} : {value}")
adder (a=2,b=4,c=3,  n_estimator=50 )
```
